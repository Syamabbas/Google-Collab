```python
# 3. Fungsi Klasifikasi
def classify_aspect(text):
    if not isinstance(text, str):
        return "Lainnya"

    text = text.lower()
    matched_aspects = []

    for aspect, keywords in aspect_keywords.items():
        for keyword in keywords:
            # Mencocokkan keyword di dalam teks
            if keyword.lower() in text:
                matched_aspects.append(aspect)
                break # Pindah ke aspek berikutnya jika sudah ditemukan kecocokan

    if not matched_aspects:
        return "Lainnya" # Jika tidak ada keyword yang cocok sama sekali

    return ", ".join(matched_aspects)
```

```python
# 4. Terapkan fungsi ke kolom clean_text
df['aspek'] = df['clean_text'].apply(classify_aspect)
```
